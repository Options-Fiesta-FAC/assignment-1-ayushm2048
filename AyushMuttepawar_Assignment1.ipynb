```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Inputs
k = float(input("Enter strike price : "))
p = float(input("Enter premium : "))
min_spot = float(input("Enter minimum spot price : "))
max_spot = float(input("Enter maximum spot price : "))
```

```python
# Generate spot prices
s= np.linspace(min_spot, max_spot, 200)
```

```python
# Calculate payoffs
buy_call = np.maximum(0, s- k ) - p 
sell_call = p - np.maximum(0, s- k )
buy_put = np.maximum(0, k - s) - p 
sell_put = p - np.maximum(0, k  - s)
# Used payoff formulas
```

```python
# Plot all payoffs
plt.plot(s, buy_call, label='Buy Call')
plt.plot(s, sell_call, label='Sell Call')
plt.plot(s, buy_put, label='Buy put')
plt.plot(s, sell_put, label='Sell put')

plt.xlabel('spot price')
plt.ylabel('profit / Loss')
plt.legend()
plt.grid(True)
plt.show()
```
